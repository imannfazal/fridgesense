```python
health_tags = [
    "low carb", "high protein", "gluten free", "low fat", "sugar free",
    "keto", "paleo", "vegan", "vegetarian", "diabetic friendly"
]

cuisine_tags = [
    "pakistani", "indian", "turkish", "chinese", "japanese", "italian",
    "mexican", "thai", "mediterranean", "arabic", "middle eastern"
]

cooking_styles = [
    "grilled", "baked", "fried", "steamed", "roasted", "slow cooked",
    "stir fried", "pressure cooked"
]
```

```python
import pandas as pd
import re

def extract_tags(text, keywords):
    found = []
    text_lower = text.lower()
    for kw in keywords:
        if re.search(rf"\b{re.escape(kw)}\b", text_lower):
            found.append(kw)
    return found

# Load your dataset
df = pd.read_csv("./data/recipes_ingredients.csv")  # change to your file path

# Create new columns
df["health_tags"] = df["description"].apply(lambda x: extract_tags(str(x), health_tags))
df["cuisine_tags"] = df["description"].apply(lambda x: extract_tags(str(x), cuisine_tags))
df["cooking_style_tags"] = df["description"].apply(lambda x: extract_tags(str(x), cooking_styles))

df.to_csv("tagged_recipes.csv", index=False)
print("Tagging complete!")
```

```python
import re

# Precompile regex for measurement parsing
MEASURE_PATTERN = re.compile(
    r'^\s*'
    r'(?P<qty>\d+(?:[\s\-]\d+\/\d+|\s\d+\/\d+|\/\d+|\.\d+|\-\d+)?|\d*\.?\d+)?'  # e.g. 1 1/2 or 3-4 or 0.5
    r'\s*'
    r'(?P<unit>(g|gram|grams|kg|ml|l|liter|liters|cup|cups|tsp|teaspoon|teaspoons|tbsp|tablespoon|tablespoons|clove|cloves|slice|slices|pinch|dash|lb|pound|pounds|oz|ounce|ounces)?)'
    r'\s*'
    r'(?P<ingredient>.+)$',
    flags=re.IGNORECASE
)

# Normalize fractions like 1 1/2 → 1-1/2 and convert to nicer form if you want
def normalize_quantity(qty):
    if not qty:
        return ""
    qty = qty.strip()
    # Replace unicode fractions with ascii if needed (optional)
    unicode_fractions = {
        "½": "1/2",
        "⅓": "1/3",
        "¼": "1/4",
        "¾": "3/4",
        # add more if needed
    }
    for uf, ascii_frac in unicode_fractions.items():
        qty = qty.replace(uf, ascii_frac)
    # Replace space fractions to 1-1/2 style
    qty = re.sub(r'(\d+)\s+(\d+/\d+)', r'\1-\2', qty)
    # Optional: convert ranges like 3-4 to 3-4 (keep as is)
    return qty

def clean_ingredient(ingredient):
    ingredient = ingredient.strip().lower()
    ingredient = re.sub(r'\s+', ' ', ingredient)  # normalize spaces
    ingredient = ingredient.strip(",.;:-")  # trim trailing punctuation
    # Parse using regex
    m = MEASURE_PATTERN.match(ingredient)
    if m:
        qty = normalize_quantity(m.group("qty"))
        unit = m.group("unit") or ""
        ing = m.group("ingredient").strip()
        # Normalize unit spacing (e.g. 250ml → 250 ml)
        if qty and unit and not qty.endswith(" "):
            qty = qty.strip()
            unit = unit.strip()
        measurement = f"{qty} {unit}".strip()
        if measurement:
            return f"{measurement}: {ing}"
        else:
            return ing
    else:
        return ingredient

def clean_ingredients_list(ingredients_list):
    if not isinstance(ingredients_list, list):
        return []
    # Use list comprehension for speed
    return [clean_ingredient(ing) for ing in ingredients_list if ing and isinstance(ing, str)]

# Apply to dataframe column (vectorized apply is ok here)
recipes_col["ingredients_cleaned"] = recipes_col["ingredients_raw"].apply(clean_ingredients_list)
```

```python
import re
import unicodedata

# Units list for recognition
UNITS = [
    "g", "gram", "grams", "kg", "kilogram", "kilograms", "ml", "milliliter", "milliliters",
    "l", "liter", "liters", "litre", "litres", "oz", "ounce", "ounces", "lb", "pound", "pounds",
    "cup", "cups", "tsp", "teaspoon", "teaspoons", "tbsp", "tablespoon", "tablespoons",
    "clove", "cloves", "slice", "slices", "pinch", "dash", "sprig", "sprigs", "can", "cans",
    "stick", "sticks", "package", "packages"
]

# Junk words to remove from ingredient text
REMOVE_WORDS = ["optional", "to taste", "as needed", "as required", "or as needed"]

# Regex for mixed fraction and ranges
FRACTION_RE = re.compile(r'(\d+)\s?[\u00BC-\u00BE\u2150-\u215E]')  # to detect mixed unicode fractions like 1½
MIXED_FRAC_RE = re.compile(r'(\d+)\s+(\d+/\d+)')  # 1 1/2
RANGE_RE = re.compile(r'(\d+(?:/\d+)?)(?:\s*-\s*|\s+to\s+)(\d+(?:/\d+)?)')  # 3-4 or 3 to 4

# Unicode fraction map to decimals for conversions
UNICODE_FRACTIONS = {
    '¼': '1/4', '½': '1/2', '¾': '3/4',
    '⅐': '1/7', '⅑': '1/9', '⅒': '1/10',
    '⅓': '1/3', '⅔': '2/3', '⅕': '1/5',
    '⅖': '2/5', '⅗': '3/5', '⅘': '4/5',
    '⅙': '1/6', '⅚': '5/6', '⅛': '1/8',
    '⅜': '3/8', '⅝': '5/8', '⅞': '7/8'
}

def replace_unicode_fractions(s):
    for uf, frac in UNICODE_FRACTIONS.items():
        s = s.replace(uf, frac)
    return s

def convert_fraction_to_decimal(frac_str):
    # Converts "1 1/2" or "3/4" to float string "1.5" or "0.75"
    try:
        if ' ' in frac_str:
            whole, fraction = frac_str.split()
            numerator, denominator = fraction.split('/')
            val = float(whole) + float(numerator) / float(denominator)
            return str(val)
        elif '/' in frac_str:
            numerator, denominator = frac_str.split('/')
            val = float(numerator) / float(denominator)
            return str(val)
        else:
            return frac_str
    except:
        return frac_str

def parse_quantity(qty_str):
    qty_str = qty_str.strip()
    qty_str = replace_unicode_fractions(qty_str)
    
    # Handle ranges by averaging
    m = RANGE_RE.match(qty_str)
    if m:
        left, right = m.groups()
        left_val = float(convert_fraction_to_decimal(left))
        right_val = float(convert_fraction_to_decimal(right))
        avg = (left_val + right_val) / 2
        return str(round(avg, 2))

    # Handle mixed fractions
    m = MIXED_FRAC_RE.match(qty_str)
    if m:
        return convert_fraction_to_decimal(qty_str)

    # Simple fraction or number
    return convert_fraction_to_decimal(qty_str)

def clean_ingredient(ing_raw):
    ing = ing_raw.strip().lower()
    
    # # Remove junk words
    # for junk in REMOVE_WORDS:
    #     pattern = r'\b' + re.escape(junk) + r'\b'
    #     ing = re.sub(pattern, '', ing)
    
    ing = ing.strip(",;:. ")

    # Normalize unicode fractions to ascii
    ing = replace_unicode_fractions(ing)

    # Extract qty + unit + ingredient with regex:
    # This pattern looks for:
    # - qty: digits, fractions, decimals, ranges
    # - optional space
    # - unit from list (allow plural 's')
    # - space + ingredient
    pattern = re.compile(
        r'^(?P<qty>[\d\s\/\.\-to]+)?\s*(?P<unit>(' + '|'.join(UNITS) + r'))?s?\s*(?P<ingredient>.+)$'
    )

    match = pattern.match(ing)
    if match:
        qty = match.group('qty') or ''
        unit = match.group('unit') or ''
        ingredient = match.group('ingredient').strip()

        qty = parse_quantity(qty) if qty else ''
        unit = unit.lower()
        
        # Handle case where qty and unit stuck together like "250ml"
        if not qty and unit:
            m2 = re.match(r'^(\d+)([a-zA-Z]+)$', ing)
            if m2:
                qty, unit = m2.groups()
                qty = parse_quantity(qty)

        # Compose output
        if qty and unit:
            cleaned = f"{qty} {unit}: {ingredient}"
        elif qty:
            cleaned = f"{qty}: {ingredient}"
        else:
            cleaned = ingredient

        # Clean extra spaces & punctuation
        cleaned = re.sub(r'\s+', ' ', cleaned)
        cleaned = cleaned.strip(",;:. ")

        # Normalize casing: first letter lowercase, rest as is
        if cleaned:
            cleaned = cleaned[0].lower() + cleaned[1:]

        return cleaned

    # fallback: just normalized ingredient string without qty/unit
    return ing

def clean_ingredients_list(ingredients_list):
    if not isinstance(ingredients_list, list):
        return []
    return [clean_ingredient(ing) for ing in ingredients_list if isinstance(ing, str) and ing.strip()]


cleaned_ingredients["ingredients_cleaned2"] = cleaned_ingredients["ingredients_raw"].apply(clean_ingredients_list)
```

```python
import re
import pandas as pd
import csv  # For quoting constants

# Map for fraction replacements
fraction_map = {
    '1/2': '½',
    '1/4': '¼',
    '3/4': '¾',
    '1/3': '⅓',
    '2/3': '⅔'
}

def replace_fractions(text):
    # Mixed numbers first: e.g., "1 1/2" -> "1 & ½"
    def mixed_number_repl(match):
        whole = match.group(1)
        frac = match.group(2)
        return f"{whole} & {fraction_map.get(frac, frac)}"
    
    text = re.sub(r'(\d+)\s+(1/2|1/4|3/4|1/3|2/3)\b', mixed_number_repl, text)

    # Single fractions (not part of mixed numbers)
    for ascii_frac, uni_frac in fraction_map.items():
        text = re.sub(rf'\b{ascii_frac}\b', uni_frac, text)

    # Ranges: e.g., "1/2-1" -> "½ to 1"
    text = re.sub(
        rf'({"|".join(map(re.escape, fraction_map.values()))}|\d+)\s*-\s*(\d+|{"|".join(map(re.escape, fraction_map.values()))})',
        lambda m: f"{m.group(1)} to {m.group(2)}",
        text
    )
    return text

def clean_ingredient(ingredient):
    # Fix lb/l/bs abbreviations
    ingredient = re.sub(r'\bl\b', 'lb', ingredient)
    ingredient = re.sub(r'\blb?s\b', 'lbs', ingredient)

    # Standardize ounce to oz
    ingredient = re.sub(r'\bounces?\b', 'oz', ingredient)
    ingredient = re.sub(r'\boz\.\b', 'oz', ingredient)

    # Expand abbreviations
    replacements = {
        r'\bpkge?\b\.?': 'package',
        r'\bpkg\b\.?': 'package',
        r'\bpk\b\.?': 'package',
    }
    for pattern, repl in replacements.items():
        ingredient = re.sub(pattern, repl, ingredient)

    # Remove stray colons after numbers
    ingredient = re.sub(r'(\d+(?:\.\d+)?)\s*:', r'\1', ingredient)

    # Fix missing space in "packageorange" etc.
    ingredient = re.sub(r'(package)([A-Za-z])', r'\1 \2', ingredient)

    # Fix number + parentheses spacing
    ingredient = re.sub(r'\s*\(\s*', ' (', ingredient)
    ingredient = re.sub(r'\s*\)\s*', ') ', ingredient)

    # Fix dots in words
    ingredient = re.sub(r'(\w)\.(\w)', r'\1 \2', ingredient)

    # Apply fraction replacements
    ingredient = replace_fractions(ingredient)

    # Clean multiple spaces
    ingredient = re.sub(r'\s+', ' ', ingredient).strip()

    return ingredient

def clean_ingredients_list(ingredients_list):
    return [clean_ingredient(ing) for ing in ingredients_list]

# # --- Example usage ---
# df = pd.read_csv("recipes.csv")

# # Convert stringified lists back to Python lists if needed
# df["ingredients_cleaned"] = df["ingredients_cleaned"].apply(eval)

# # Apply cleaning
# df["ingredients_cleaned"] = df["ingredients_cleaned"].apply(clean_ingredients_list)

# # Save with safe quoting so commas/quotes in ingredients don't break CSV
# df.to_csv("ingredients_cleaned.csv", index=False, quoting=csv.QUOTE_ALL)
```

```python
import pandas as pd

nonveg_keywords = [
    "beef", "pork", "chicken", "fish", "lamb", "mutton",
    "shrimp", "prawn", "crab", "lobster", "bacon", "ham",
    "sausage", "anchovy", "turkey", "duck", "veal", "goat", "meat"
]

def contains_nonveg(ingredients):
    if pd.isna(ingredients):
        return False
    ing_list = str(ingredients).lower()
    return any(word in ing_list for word in nonveg_keywords)

def fix_tags(row):
    if pd.isna(row["merged_str"]):
        return row["merged_str"]  # leave NaN as is
    if row["merged_str"].strip().lower() == "vegetarian" and contains_nonveg(row["ingredients"]):
        return "non-vegetarian"
    return row["merged_str"]

half_done["merged_str"] = half_done.apply(fix_tags, axis=1)
```

```python
import pandas as pd
import re

def merge_tag_columns(row):
    # Convert tuple-like strings into lists
    tags_1 = str(row["clean_tags_tuple"]).strip("()").replace("'", "").split(",")
    tags_2 = str(row["merged_str"]).strip("()").replace("'", "").split(",") if pd.notna(row["merged_str"]) else []
    
    merged = []
    seen_lower = set()
    
    for tag in tags_1 + tags_2:
        tag = tag.strip()
        
        # Skip empty tags
        if not tag:
            continue
        
        # Remove tags ending with -<number> (e.g., healthy-2)
        if re.search(r"-\d+$", tag):
            continue
        
        tag_lower = tag.lower()
        
        # Deduplicate, preferring uppercase if it exists
        if tag_lower not in seen_lower:
            merged.append(tag.upper() if tag.isupper() else tag)
            seen_lower.add(tag_lower)
    
    return ", ".join(merged)

half_done["tags_merged"] = half_done.apply(merge_tag_columns, axis=1)
```

```python
import re
import pandas as pd

# 1️⃣ Remove duplicate IDs (keep the first occurrence)
done = done.drop_duplicates(subset=['id'], keep='first')

# 2️⃣ Strip leading/trailing whitespace from all string columns
str_cols = done.select_dtypes(include=['object']).columns
for col in str_cols:
    done[col] = done[col].str.strip()

# 3️⃣ Remove non-printable characters from all string columns
pattern = re.compile(r'[^\x20-\x7E]')  # ASCII printable range
for col in str_cols:
    done[col] = done[col].apply(
        lambda x: pattern.sub(' ', x) if isinstance(x, str) else x
    )

# 4️⃣ Ensure tags_merged has only unique, uppercase tags
def clean_tags(tags):
    if isinstance(tags, str):
        unique_tags = sorted(set(tags.split(', ')), key=str.upper)
        # Keep only uppercase versions if duplicate in upper/lower
        final_tags = []
        seen_upper = set()
        for t in unique_tags:
            if t.upper() not in seen_upper:
                seen_upper.add(t.upper())
                final_tags.append(t.upper())
        return ', '.join(final_tags)
    return tags

done['tags_merged'] = done['tags_merged'].apply(clean_tags)

# 5️⃣ Reset index after cleaning
done.reset_index(drop=True, inplace=True)

print("Deep cleanse complete ✅")
```
